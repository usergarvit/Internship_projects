```python
#Conversion kg to pounds:

kg = float(input("Enter weight in kilograms: "))
pounds = kg * 2.20462
print("Pounds:", pounds)

```

```python
#Conversion inr to usd:

inr = float(input("Enter amount in rupees: "))
usd = inr / 83
print("Dollars:", usd)
```

```python
#Conversion cm to inch:

cm = float(input("Enter height in cm: "))
inch = cm / 2.54
print("Inches:", inch)
```

```python
#Conversion minutes to seconds:

minutes = float(input("Enter minutes: "))
sec = minutes * 60
print("Seconds:", sec)

```
