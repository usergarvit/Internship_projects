```python
import pyttsx3
import os
import webbrowser
import speech_recognition as sr
import pywhatkit


speaker = pyttsx3.init('sapi5')


mic = sr.Recognizer()

speaker.say("Hello, I am Jenny. How can I help you?")
speaker.runAndWait()

def listen_command():
    with sr.Microphone() as source:
        print("\n Speak your command...")
        mic.adjust_for_ambient_noise(source, duration=1)

        try:
            voice = mic.listen(source, timeout=5, phrase_time_limit=5)
            text = mic.recognize_google(voice)
            return text.lower()

        except sr.WaitTimeoutError:
            print(" No speech detected.")
            return ""

        except sr.UnknownValueError:
            print(" Could not understand.")
            return ""

        except sr.RequestError:
            print("Speech service is unavailable.")
            return ""


while True:   
    command = listen_command()

    if command == "":
        continue  
    print("You said:", command)

   
    if "stop" in command or "exit" in command or "bye" in command:
        print(" Bye!")
        speaker.say("Goodbye.")
        speaker.runAndWait()
        break

   
    if "notepad" in command or "note pad" in command:
        print(" Opening Notepad...")
        speaker.say("Opening Notepad")
        speaker.runAndWait()
        os.system("notepad")

    elif "youtube" in command or "you tube" in command or "yt" in command:
        print(" Opening YouTube...")
        speaker.say("Opening YouTube")
        speaker.runAndWait()
        webbrowser.open("https://www.youtube.com")

    elif "spotify" in command or "spot fy" in command:
        print(" Opening Spotify...")
        speaker.say("Opening Spotify")
        speaker.runAndWait()
        try:
            os.system("start spotify")
        except:
            webbrowser.open("https://open.spotify.com")
            print("Spotify desktop not found. Opening web version.")

    elif "whatsapp" in command or "send message" in command:
        print(" Sending WhatsApp message...")
        speaker.say("Sending WhatsApp message")
        speaker.runAndWait()
        try:
            pywhatkit.sendwhatmsg_instantly("+919413722057", "Hi, I am Jenny!", wait_time=10, tab_close=True)
            print("Message sent!")
        except Exception as e:
            print("Error:", e)

    else:
        print(" Command not recognized.")
        speaker.say("Sorry, I did not understand.")
        speaker.runAndWait()

speaker.stop()
```
